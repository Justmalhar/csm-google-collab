```python
# Load the model
from generator import load_csm_1b
import torchaudio

generator = load_csm_1b(device="cuda")
print("Model loaded successfully!")
```

```python
# Create a function to generate audio
def generate_audio(text, speaker=0):
    # Generate audio from text
    audio = generator.generate(
        text=text,
        speaker=speaker,
        context=[],
        max_audio_length_ms=10_000,
    )

    # Save the generated audio
    output_file = "output_audio.wav"
    torchaudio.save(output_file, audio.unsqueeze(0).cpu(), generator.sample_rate)

    return output_file
```

```python
# Define the Gradio interface function
def gradio_interface(text, speaker):
    try:
        # Call the generate_audio function
        audio_file = generate_audio(text, speaker)
        return audio_file
    except Exception as e:
        print(f"Error generating audio: {e}")
        return None
```

GRADIO INTERFACE

```python
import gradio as gr

# Define the Gradio interface using Blocks API
with gr.Blocks(title="Sesame CSM-1B Text-to-Speech") as demo:
    gr.Markdown("# 🎙️ Sesame CSM-1B Text-to-Speech")
    gr.Markdown("Generate high-quality audio from text using the Sesame CSM-1B model.")

    with gr.Row():
        with gr.Column():
            # Text input
            text_input = gr.Textbox(label="Enter Text", placeholder="Type your text here...", lines=5)

            # Speaker selection
            speaker_input = gr.Number(
                label="Speaker ID (0-1000)",
                value=0,
                minimum=0,
                maximum=10,
                interactive=True
            )

            # File upload for text
            file_upload = gr.File(label="Or Upload a Text File", file_types=[".txt"])

            # Generate button
            generate_button = gr.Button("Generate Audio 🎵")

        with gr.Column():
            # Audio output
            audio_output = gr.Audio(label="Generated Audio", interactive=False)

            # Playback controls
            with gr.Row():
                play_button = gr.Button("▶️ Play")
                pause_button = gr.Button("⏸️ Pause")
                stop_button = gr.Button("⏹️ Stop")

            # Volume control
            volume_slider = gr.Slider(minimum=0, maximum=100, value=50, label="Volume")

    # Store the audio state
    audio_state = gr.State(value=None)  # Stores the current audio file path

    # Define interactions
    def process_input(text, file, speaker_id):
        if file is not None:
            # Read text from the uploaded file
            with open(file.name, "r") as f:
                text = f.read()
        # Ensure speaker_id is within the valid range (0-1000)
        speaker_id = int(speaker_id)  # Convert to integer
        speaker_id = max(0, min(speaker_id, 1000))  # Clamp to range

        # Generate audio (replace this with your actual audio generation logic)
        audio_file = generate_audio(text, speaker_id)
        return audio_file, audio_file  # Return audio file for output and state

    # Play audio
    def play_audio(audio_state):
        if audio_state:
            return audio_state
        return None

    # Pause audio
    def pause_audio(audio_state):
        if audio_state:
            return None  # Pausing is handled by the frontend
        return None

    # Stop audio
    def stop_audio(audio_state):
        if audio_state:
            return None  # Stopping is handled by the frontend
        return None

    # Link inputs and outputs
    generate_button.click(
        fn=process_input,
        inputs=[text_input, file_upload, speaker_input],
        outputs=[audio_output, audio_state]
    )

    # Link playback controls
    play_button.click(
        fn=play_audio,
        inputs=audio_state,
        outputs=audio_output
    )
    pause_button.click(
        fn=pause_audio,
        inputs=audio_state,
        outputs=audio_output
    )
    stop_button.click(
        fn=stop_audio,
        inputs=audio_state,
        outputs=audio_output
    )

# Launch the Gradio app
demo.launch(share=True)  # Set `share=True` to get a public link
```
